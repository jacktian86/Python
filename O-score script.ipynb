```python
from pandas import DataFrame, date_range, read_csv, read_excel, concat, isnull
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy.random as np
import ols
import statsmodels.api as sm
import math

from datetime import datetime, timedelta

%matplotlib inline
```

```python
df = read_excel('Book2.xlsx', 'Sheet1')
df = df.rename(columns={"Liabilities To Assets - Lookup": "Liability/Asset Ratio"})
df.info()

df = df[df['Date']>5]
df_date_match = df[['CIF', 'Date']]
df_date_match['CIF']=df_date_match['CIF'].apply(str)

df_date_match['CIF']= [item.strip('0.') for item in df_date_match['CIF']]
df_date_match = df_date_match.groupby(['CIF']).mean()

df=df.dropna(axis=0)
df=df[df['Liability/Asset Ratio']>.1]

index_var = 'CIF'

df2 = df[[index_var, 'Net Profit', 'Net Sales', 'Retained Earnings', 'EBITDA', 'Total Current Liabilities', 'Total Current Assets', 'Liability/Asset Ratio', 'Total Assets']]
df2['Total Liabilities'] = df2['Total Assets']*df2['Liability/Asset Ratio']
df2[index_var]=df2[index_var].apply(str)
df2[index_var]= [item.strip('0.') for item in df2[index_var]] 

df = df2.groupby([index_var]).sum()

def Z_upper (x=0):
    Upper_lim = (x.quantile(q=.75)-x.quantile(q=.5))*2+x.quantile(q=.5)
    return Upper_lim

def Z_lower (x=0):
    Lower_lim = (x.quantile(q=.25)-x.quantile(q=.5))*2+x.quantile(q=.5)
    return Lower_lim

T1 = pd.Series([math.log(item,10) for item in df['Total Assets']], index=df.index)
T2 = df['Total Liabilities']/df['Total Assets']
T3 = (df['Total Current Assets']-df['Total Current Liabilities'])/df['Total Assets']
T4 = df['Total Current Liabilities']/df['Total Current Assets']
T5 = pd.Series([1 if item>1 else 0 for item in T4], index = df.index)

T6 = df['Net Profit']/df['Total Assets']
T6 = pd.Series([max(item, 0) for item in T6], index = df.index)
T7 = df['EBITDA']/df['Total Liabilities']
T7 = pd.Series([max(item, 0) for item in T7], index = df.index)
T8 = pd.Series([1 if item==0 else 0 for item in T6*T7], index = df.index)

mask1 = pd.Series([item>Z_lower(T1) and item<Z_upper(T1) for item in T1], index = df.index)
mask2 = pd.Series([item>Z_lower(T2) and item<Z_upper(T2) for item in T2], index = df.index)
mask3 = pd.Series([item>Z_lower(T3) and item<Z_upper(T3) for item in T3], index = df.index)
mask4 = pd.Series([item>Z_lower(T4) and item<Z_upper(T4) for item in T4], index = df.index)

mask6 = pd.Series([item>Z_lower(T6) and item<Z_upper(T6) for item in T6], index = df.index)
mask7 = pd.Series([item>Z_lower(T7) and item<Z_upper(T7) for item in T7], index = df.index)

T1 = T1[mask1]
T2 = T2[mask2]
T3 = T3[mask3]
T4 = T4[mask4]

T6 = T6[mask6]
T7 = T7[mask7]

dfx = DataFrame({'T1':T1,'T2':T2,'T3':T3,'T4':T4,'T5':T5,'T6':T6,'T7':T7,'T8':T8})
dfx = dfx.dropna(axis = 0)
Z = .407*dfx['T1']-6.03*dfx['T2']+1.43*dfx['T3']+-.757*dfx['T4']+2.37*dfx['T5']+1.83*dfx['T6']-.285*dfx['T7']+1.72*dfx['T8']
dfx['Z'] = Z

```

```python
fig1 = plt.figure(1)

ax1 = fig1.add_subplot(241)
ax1.plot(range(len(dfx)), dfx['T1'], 'b')

ax2 = fig1.add_subplot(242)
ax2.plot(range(len(dfx)), dfx['T2'], 'b')

ax3 = fig1.add_subplot(243)
ax3.plot(range(len(dfx)), dfx['T3'], 'b')

ax4 = fig1.add_subplot(244)
ax4.plot(range(len(dfx)), dfx['T4'], 'b')

ax5 = fig1.add_subplot(245)
ax5.plot(range(len(dfx)), dfx['T5'], 'b')

ax6 = fig1.add_subplot(246)
ax6.plot(range(len(dfx)), dfx['T6'], 'b')

ax7 = fig1.add_subplot(247)
ax7.plot(range(len(dfx)), dfx['T7'], 'b')

ax8 = fig1.add_subplot(248)
ax8.plot(range(len(dfx)), dfx['T8'], 'b')

fig2 = plt.figure(2)
ax1 = fig2.add_subplot(111)
ax1.plot(range(len(dfx)), dfx['Z'])
```

```python
df_regress = read_excel('Calcs.xlsx', 'Sheet1')

df_regress['Customer_Nbr_Txt']=df_regress['WMS Customer .']
df_regress['Customer_Nbr_Txt']=df_regress['Customer_Nbr_Txt'].apply(str)
df_regress['Customer_Nbr_Txt']=[item.strip('0') for item in df_regress['Customer_Nbr_Txt']] 
df_regress_new = df_regress.groupby(['Customer_Nbr_Txt']).mean()
df_regress_new['Days Aged - Past Due Interest and Fees 2']=df_regress_new['Days Aged - Past Due Interest and Fees 2'].fillna(0)

max_regress1 = df_regress_new.iloc[:,-5::]
max_regress1['Risk Rating'] = df_regress_new['Risk Rating (1-9)']
max_regress1 = max_regress1.dropna(axis=0)
max_regress1.info()
df_regress2=read_excel('IntelliCheck Data_by Week 2 13 2014.xlsx', 'Sheet1')

df_regress2['CARE Number'] = df_regress2['CARE Number'].apply(str)
df_regress2['CARE Number']= [item.strip('0.') for item in df_regress2['CARE Number']] 
df_regress2['Credit Ratio'] = df_regress2['Monthly Beg Balance']/df_regress2['Credit Limit']

holder = 100
Date_Holder= pd.Series([holder if item not in df_date_match.index else df_date_match.ix[item, 'Date'] for item in df_regress2['CARE Number']], index=df_regress2.index)

df_regress2['Date Holder'] = Date_Holder
df_regress2 = df_regress2[df_regress2['Date Holder']<holder]
df_regress2 = df_regress2[df_regress2['Rolling_28_Day']<=df_regress2['Date Holder']]

Metric_Analysis = df_regress2.ix[:,['CARE Number', 'Weekly Percent Paid', 'Monthly Percent Paid', 'Credit Ratio']]
Vol_index = Metric_Analysis.groupby('CARE Number').std()
Vol_index=Vol_index.rename(columns = {'Credit Ratio':'CR Vol', 'Monthly Percent Paid':'Monthly Vol', 'Weekly Percent Paid':'Weekly Vol'})

Metric_Analysis = df_regress2.ix[:,['CARE Number', 'Weekly Percent Paid', 'Monthly Percent Paid', 'Credit Ratio']]
Mean_index = Metric_Analysis.groupby('CARE Number').mean()
Metrics_combined = Vol_index.join(Mean_index, how='inner')
```

```python
df_combined = dfx.join(Metrics_combined, how = 'inner')
df_combined = df_combined[df_combined['Credit Ratio']<=1]
df_combined = df_combined.dropna(axis=0)
```

```python
regress = df_combined.ix[:, ['Monthly Percent Paid', 'Credit Ratio', 'Weekly Vol']]
regress = sm.add_constant(regress)

model = sm.OLS(df_combined['Z'], regress)
fit = model.fit()
alpha=fit.params

fit.summary()
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(121)
ax1.plot(df_combined['Z'], fit.fittedvalues, 'g.',)

ax2 = fig1.add_subplot(122)
ax2.plot(range(len(df_combined)), df_combined['Z'], 'b.')
```
